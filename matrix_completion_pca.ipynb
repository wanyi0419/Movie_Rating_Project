```python
import pandas as pd 
import numpy as np
import csv
import math
```

```python
user_history = pd.read_csv("user_history.csv")
user_history
```

```python
user_ratings = pd.read_csv("user_ratings.csv")
user_ratings_table = pd.pivot_table(user_ratings,values='RATING', index='USER ID',
                    columns='PRODUCT')
train_rating = user_ratings_table.sample(frac = 0.8)
test_rating = user_ratings_table.drop(train_rating.index)
```

```python
#Doing PCA (in order to do PCA we must strip the first column then transpose our matrix)
X = user_history.loc[:, user_history.columns != 'USER ID'].to_numpy()
X_meanzero = (X - np.mean(X, axis = 0)) # subtract the mean 
X = (X_meanzero / np.std(X, axis = 0)) # divide by the standard deviation
U,S,Vt = np.linalg.svd(X)
# Compute the fraction of the total spectrum in the first 20 singular values
FractionFromS1_S20 = sum(S[0:20]**2) / sum(S**2)
print("Fraction of Spectrum from sigma's 0 through 20 =", FractionFromS1_S20)
```

```python
#Now let us use this top 20 singular vectors to make a matrix 
user_data_t20 = pd.DataFrame(data = U[:,0:20], index = user_history['USER ID'])
user_data_t20
```

```python
#Now lets make the matrix that we need to complete: the training values attached to their corresponding user ids
user_data_with_ratings_incomplete = user_data_t20.join(train_rating, how = 'right')
user_data_with_ratings_incomplete
```
